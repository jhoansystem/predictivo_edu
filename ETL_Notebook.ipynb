Este cuaderno realiza de forma detallada la extracción del dataset original, la limpieza de valores atípicos/nulos, transformaciones de tipos de datos, creación de variables predictoras y almacenamiento del archivo limpio final listo para el modelado.

```python
# Paso 1: Importación de librerías esenciales preinstaladas en Google Colab
import pandas as pd
import numpy as np
print("Librerías cargadas exitosamente.")
```

### Paso 2: Extracción de Datos (Extract)
Cargamos el archivo CSV original utilizando el separador de punto y coma (`;`) típico de las exportaciones locales.

```python
# Carga del archivo original
path_archivo = 'Recursos/Resultados 2015 - 2024.csv'
df_raw = pd.read_csv(path_archivo, sep=';')

# Visualización preliminar de dimensiones
print(f"Dimensiones iniciales del dataset: {df_raw.shape[0]} filas y {df_raw.shape[1]} columnas.\n")
print("Primeras 3 filas del archivo original:")
df_raw.head(3)
```

### Paso 3: Transformación y Limpieza profunda (Transform)
1. Identificamos los strings `'NULL'` de la base de datos y los cambiamos por `np.nan` real.
2. Convertimos las notas a flotantes numéricos.
3. Rellenamos faltas nulas con `0` y tipificamos como enteros.
4. Creamos la columna binaria `target_binario` (1 = Aprobado/Promovido 'S', 0 = Reprobado 'N').

```python
df_clean = df_raw.replace('NULL', np.nan)

df_clean['nota'] = pd.to_numeric(df_clean['nota'], errors='coerce')
df_clean['faltas'] = pd.to_numeric(df_clean['faltas'], errors='coerce').fillna(0).astype(int)
df_clean['intensidad_semannal'] = pd.to_numeric(df_clean['intensidad_semannal'], errors='coerce').fillna(0).astype(int)
df_clean['grado'] = pd.to_numeric(df_clean['grado'], errors='coerce').fillna(0).astype(int)
df_clean['refuerzo'] = pd.to_numeric(df_clean['refuerzo'], errors='coerce').fillna(0).astype(int)

df_clean['target_binario'] = df_clean['promovido'].apply(lambda x: 1 if str(x).strip().upper() == 'S' else 0)

columnas_interes = ['ano', 'grado', 'intensidad_semannal', 'faltas', 'nota', 'refuerzo', 'target_binario']
df_final = df_clean[columnas_interes].dropna()

print(f"Dimensiones después de la limpieza y filtrado: {df_final.shape}")
df_final.head()
```

### Paso 4: Carga de Datos Procesados (Load)
Guardamos el set de datos limpio en un nuevo archivo CSV para alimentar el EDA y el modelado predictivo de forma independiente.

```python
# Exportar el archivo definitivo
df_final.to_csv('Resultados_Limpio.csv', index=False)
print("¡Archivo 'Resultados_Limpio.csv' exportado correctamente listo para el EDA y Modelado!")
```
