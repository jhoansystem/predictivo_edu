# 📓 Cuaderno 3: Ingeniería y Evaluación de Modelos de ML
**Proyecto:** Predicción de Éxito Académico y Mitigación de Deserción Escolar
**Metodología:** CRISP-ML - Fases de Modelado y Evaluación

En este cuaderno entrenaremos paso a paso tres modelos analíticos:
1. **Regresión Lineal** para predecir el valor continuo de la `nota` final.
2. **Regresión Logística** para estimar la probabilidad de aprobación (`target_binario`).
3. **Árbol de Decisión** para generar reglas interpretables sobre el éxito escolar.

```python
# Paso 1: Importar los algoritmos matemáticos y métricas de scikit-learn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import r2_score, mean_squared_error, accuracy_score, classification_report, confusion_matrix
import joblib
import pandas as pd
import numpy as np
print("Entorno de Modelado Inteligente Inicializado.")
```

### Paso 2: Carga y Partición de Datos (Train/Test Split)
Separamos el dataset en un 80% para entrenamiento de la IA y un 20% para la validación ciega de su desempeño.

```python
df = pd.read_csv('Resultados_Limpio.csv')

# Definición de características predictoras (X) y variables objetivo (y)
X_clasif = df[['grado', 'intensidad_semannal', 'faltas', 'nota' , 'refuerzo']]
y_clasif = df['target_binario']

X_train, X_test, y_train, y_test = train_test_split(X_clasif, y_clasif, test_size=0.2, random_state=42)
print(f"Set de Entrenamiento: {X_train.shape[0]} filas | Set de Prueba: {X_test.shape[0]} filas.")
```

### Paso 3: Entrenamiento de Modelo 1 - Regresión Lineal
Predecimos el valor continuo de la `nota` basándonos en grado, intensidad semanal, procesos de refuerzo y faltas históricas.

```python
X_lin = df[['grado', 'intensidad_semannal', 'faltas' , 'refuerzo']]
y_lin = df['nota']

X_train_l, X_test_l, y_train_l, y_test_l = train_test_split(X_lin, y_lin, test_size=0.2, random_state=42)

modelo_lineal = LinearRegression()
modelo_lineal.fit(X_train_l, y_train_l)
preds_lin = modelo_lineal.predict(X_test_l)

print("--- Evaluación Regresión Lineal ---")
print(f"Coeficiente de Determinación R² Score: {r2_score(y_test_l, preds_lin):.4f}")
print(f"Error Cuadrático Medio (MSE): {mean_squared_error(y_test_l, preds_lin):.4f}")
```

### Paso 4: Entrenamiento de Modelo 2 - Regresión Logística
Modelo probabilístico clásico de clasificación binaria.

```python
modelo_logistico = LogisticRegression(max_iter=1000)
modelo_logistico.fit(X_train, y_train)
preds_log = modelo_logistico.predict(X_test)

print("--- Evaluación Regresión Logística ---")
print(f"Exactitud (Accuracy): {accuracy_score(y_test, preds_log):.4f}\n")
print(classification_report(y_test, preds_log))
```

### Paso 5: Entrenamiento de Modelo 3 - Árbol de Decisión
Generación de un mapa de decisiones jerárquicas lógicas altamente interpretables por docentes.

```python
modelo_arbol = DecisionTreeClassifier(max_depth=4, random_state=42)
modelo_arbol.fit(X_train, y_train)
preds_tree = modelo_arbol.predict(X_test)

print("--- Evaluación Árbol de Decisión ---")
print(f"Exactitud (Accuracy): {accuracy_score(y_test, preds_tree):.4f}\n")
print(classification_report(y_test, preds_tree))
```

### Paso 6: Serialización y Exportación de Modelos para Producción
Exportamos los archivos `.pkl` binarios binarios que usará la aplicación Streamlit para interactuar en tiempo real en la web.

```python
joblib.dump(modelo_lineal, 'modelo_lineal.pkl')
joblib.dump(modelo_logistico, 'modelo_logistico.pkl')
joblib.dump(modelo_arbol, 'modelo_arbol.pkl')
print("¡Modelos guardados con éxito en archivos locales .pkl!")
```
